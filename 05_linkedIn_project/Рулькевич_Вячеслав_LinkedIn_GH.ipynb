Создадим функцию для извлечения навыков между тэгами - 

```python
# Функция для извлечения информации между тегами
def extract_skills(value):
    start_tag = '<span class="visually-hidden"><!-- -->Skills:'
    end_tag = '<!-- --></span>'
    start_index = value.find(start_tag) + len(start_tag)
    end_index = value.find(end_tag)
    return value[start_index:end_index]

# Применение функции к столбцу с использованием метода apply()
df_da['new_skills'] = df['skills'].apply(extract_skills)
```

Удалим из списка лишние элементы,  то что не относится к Hard-скиллам

```python
unsorted = ('',' +8 more',' +7 more',' +6 more',' +5 more',' +4 more',' +3 more',' +2 more',' +1 more',
           ' ключевые показатели деятельности',' коммуникация',' наука о данных',' панель управления',
            ' решение задач',' социальные медиа',' основные показатели',' компьютерная наука',' аналитика',
            ' истории пользователей',' командная работа',' статистика по компании'
           
           )
for element in unsorted:
    top_skills.remove(element)
```
